```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fcc-forum-pageviews.csv', index_col='date', parse_dates=True)
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df.index, df['value'], color='red', linewidth=1.5
            , label='Page Views')
ax.set_title('Daily freeCodeCamp Forum Page Views 5/2016-12/2019')
ax.set_xlabel('Date')
ax.set_ylabel('Page Views')
```

```python
df_bar = pd.read_csv('fcc-forum-pageviews.csv', index_col='date', parse_dates=True)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
df_bar['year'] = df_bar.index.year
df_bar['month'] = df_bar.index.month_name()
ax.legend(title='Months')
sns.barplot(x='year', y='value', hue='month', data=df_bar)
```

```python
df = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Gender': ['Male', 'Female'] * 3,
    'Values': [10, 12, 20, 22, 15, 17]
})

# Grouped bar plot
sns.barplot(x='Category', y='Values', hue='Gender', data=df)
plt.title("Grouped Bar Plot")
plt.show()
```

```python
df_box = df.copy()
df_box.reset_index(inplace=True)
df_box['year'] = [d.year for d in df_box.date]
df_box['month'] = [d.strftime('%b') for d in df_box.date]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
sns.boxplot(x='year', y='value', data=df_box, ax=ax1
        , palette='viridis')
ax1.set_title('Year-wise Box Plot (Trend)')

sns.boxplot(x='month', y='value', data=df_box, ax=ax2
            , palette='viridis')
ax2.set_title('Month-wise Box Plot (Seasonality)')
```
